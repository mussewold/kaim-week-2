# Telecom User Overview Analysis

This notebook contains the analysis of telecom user behavior and handset usage patterns.

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scripts.data_processing_utils import *
from scripts.analysis_utils import *
from scripts.visualization_utils import *
```

## Task 1 - Handset Analysis

```python

# Load and preprocess the data
df = load_and_preprocess_data()

print("Dataset Overview:")
print("-----------------")
print(f"Number of records: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
print("\nColumns in dataset:")
print(df.columns.tolist())
print("\nSample of the data:")
print(df.head())

# Perform handset analysis
print("\nTop 10 Handsets:")
print(get_top_handsets(df))

print("\nTop 3 Manufacturers:")
top_manufacturers = get_top_manufacturers(df)
print(top_manufacturers)

print("\nTop 5 Handsets per Manufacturer:")
top_handsets_by_manufacturer = get_top_handsets_per_manufacturer(df, top_manufacturers.index.tolist())
for manufacturer, handsets in top_handsets_by_manufacturer.items():
    print(f"\n{manufacturer}:")
    print(handsets)

# Aggregate user metrics
user_metrics = aggregate_user_metrics(df)
print("\nUser Metrics Summary:")
print(user_metrics.describe())
```

## Task 1.1 - User Behavior Analysis

```python
# Aggregate user metrics
user_metrics = aggregate_user_metrics(df)

# Handle missing values
user_metrics = handle_missing_values(user_metrics)

# Create decile classes
user_metrics = create_decile_classes(user_metrics, 'duration')

print("User Metrics Summary:")
print(user_metrics.describe())
```

## Task 1.2 - Exploratory Data Analysis

```python
# Univariate Analysis
numeric_columns = user_metrics.select_dtypes(include=[np.number]).columns

for column in numeric_columns:
    print(f"\nAnalysis for {column}:")
    print(user_metrics[column].describe())
    plot_distribution(user_metrics, column)
    plot_boxplot(user_metrics, column)

# Correlation Analysis
app_columns = ['social_media_data', 'google_data', 'email_data', 'youtube_data',
               'netflix_data', 'gaming_data', 'other_data']
correlation_matrix = compute_correlation_matrix(user_metrics, app_columns)
plot_correlation_matrix(correlation_matrix)

# PCA Analysis
pca, pca_result = perform_pca(user_metrics, app_columns)
plot_pca_explained_variance(pca)

print("\nPCA Components Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```
